```python
import os
m = {}
suits = ['club','spade','diamond','heart']
card_names = ['Ace','2','3','4','5','6','7','8','9','10','Jack','Queen','King']
for f in os.listdir('./images/'):
    name = f[:-4].split('_') #remove .png
    num, suit = name[0], name[-1]
    # m[(num,suit)] = Image(f'images/{f}',height=220,width=110)
    m[(num,suit)] = image.open(f'images/{f}')
# for i in range(2,5):
#     display(m[(f'{i}','hearts')])

def get(n,s):
    return m[(f'{n}',s)]

l = [get(3,'spades'),get(6,'hearts')]


import ipyplot

_ = ipyplot.plot_images(l, max_images=20, img_width=150)
```
